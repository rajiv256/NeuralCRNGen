```julia
function unravel_params(params; dims=3)
    d2 = dims^2 
    theta = params[1:d2]
    theta = reshape(theta, (dims, dims))
    beta = params[d2 + 1 : d2 + dims]
    w = params[d2 + dims + 1 : d2 + 2*dims]
    return theta, beta, w
end
```

```julia
DIMS=3
params_orig = create_node_params(DIMS, t0=0.0, t1=1.0)
for i in 1:length(params_orig)
    params_orig[i] = abs(params_orig[i])
end

theta, beta, w, h, t0, t1 = sequester_params(params_orig, dims=3)
x = [1, 0.5, 0.2]
```

```julia
function f(theta, beta, z, x, h, t)
    hvec = ones(length(z))*h
    fmat = (theta*x + beta).*z - z.*z + hvec
    return fmat
end

function forward!(du, u, xAndp, t; dims=3)
    x = xAndp[1:dims]
    p = xAndp[dims+1:end]
    theta, beta, w, h, _, _ = sequester_params(p)
    fmat = f(theta, beta, u, x, h, t)
    for i in 1:length(fmat)
        du[i] = fmat[i]
    end
end

function backward!(du, u, sAndp, t)
    s = sAndp[1:3]
    p = sAndp[4:end]
    theta, beta, w, h, _, _ = sequester_params(p)
    fmat = f(theta, beta, u, x, h, t)
    for i in 1:length(fmat)
        du[i] = fmat[i]
    end
end

u0 = x
p = params_orig
xAndp = []
append!(xAndp, u0)
append!(xAndp, p)
```

```julia
relu.(theta*x + beta)
```

```julia
tspan = (0.0, 2.0)
fwdODE = ODEProblem(forward!, u0, tspan, xAndp)
sol = solve(fwdODE, Tsit5(), reltol=1e-8, abstol=1e-8)
print(sol[end])
plot(sol)
```

```julia
s0 = sol[end]
rtspan = (2.0, 0.0)
sAndp = []
append!(sAndp, s0)
append!(sAndp, p)
bwdODE = ODEProblem(backward!, s0, rtspan, sAndp)
solbwd = solve(bwdODE, Tsit5(), reltol=1e-8, abstol=1e-8)
print(solbwd[end])
plot(solbwd)
```
